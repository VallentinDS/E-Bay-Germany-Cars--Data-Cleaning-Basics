```python
autos["brand"].describe()
```

There is an occurence where there is a brand called `sonstige_autos` but from my research this represents other type of brands that may not be relevant or the owner did not tag his car with a brand name.

Since this is not relevant to my analysis in finding data about specific brands, I will drop off the rows with these tags.

```python
find_row= autos[autos["brand"] == "sonstige_autos"].index
autos= autos.drop(find_row)
```

```python
autos["brand"].value_counts(normalize= True) * 100
```

```python
auto_brands= autos["brand"].value_counts().head(10).index
auto_brands
```

```python
brand_mean_prices= {}

for b in auto_brands:
    selected_rows= autos[autos["brand"] == b]
    sorted_rows= selected_rows.sort_values("price", ascending= False)
    mean= sorted_rows["price"].mean().round()
    brand_mean_prices[b] = int(mean)

brand_mean_prices
```

I can see that Mini, Audi, Mercedes Benz and BMW are among the most expensive cars, while at the other end there is Renault, Fiat, Peugeot and Opel.

The most popular car brand, Volkswagen, is in between as price making it an affordable car for medium-income people.

```python
brand_mean_mileage= {}

for b in auto_brands:
    selected_rows= autos[autos["brand"] == b]
    sorted_rows= selected_rows.sort_values("odometer_km", ascending= False)
    mean= sorted_rows["odometer_km"].mean().round()
    brand_mean_mileage[b]= int(mean)
    
brand_mean_mileage
```

```python
bmp_series= pd.Series(brand_mean_prices)
bmp_series
```

```python
bmp_df= pd.DataFrame(bmp_series, columns= ["mean_price"])
bmp_df
```

```python
bmm_series= pd.Series(brand_mean_mileage)
bmp_df["mean_mileage"]= bmm_series
```
